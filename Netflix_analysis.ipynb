```python
import pandas as pd

df = pd.read_csv('../data/netflix_titles.csv')

df.head()
```

```python
movies = df[df['type'] == 'Movie'].copy().reset_index(drop=True)
movies.shape
```

```python
movies[['title', 'description']].head()
```

```python
# Fill any missing descriptions with empty string
movies['description'] = movies['description'].fillna('')

# (Optional but useful later) also include genre in text
movies['text'] = movies['description'] + ' ' + movies['listed_in'].fillna('')
movies[['title', 'text']].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['text'])

tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
cosine_sim.shape
```

```python
# Reset index to be safe
movies = movies.reset_index(drop=True)

# Create a mapping from movie title to its index
title_to_index = {title: idx for idx, title in enumerate(movies['title'])}

len(title_to_index)
```

```python
def recommend(movie_title, num_recommendations=5):
    movie_title = movie_title.strip()

    if movie_title not in title_to_index:
        return f"❌ Movie not found: {movie_title}"

    idx = title_to_index[movie_title]

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # skip the first because it will be the same movie
    sim_scores = sim_scores[1:num_recommendations+1]

    recommended_titles = [movies['title'][i] for i, _ in sim_scores]

    return recommended_titles
```

```python
recommend("The Matrix")
```

```python
def show_recommendations(movie_title, num_recommendations=5):
    result = recommend(movie_title, num_recommendations)

    if isinstance(result, str):
        print(result)
        return

    print(f"\n🎬 Because you watched: {movie_title}\n")
    for i, title in enumerate(result, start=1):
        print(f"{i}. {title}")
```

```python
show_recommendations("The Matrix", 5)
```

```python
# Safer fills
movies['description'] = movies['description'].fillna('')
movies['listed_in'] = movies['listed_in'].fillna('')
movies['cast'] = movies['cast'].fillna('')
movies['director'] = movies['director'].fillna('')

# Richer text field
movies['text'] = (
    movies['description'] + ' ' +
    movies['listed_in'] + ' ' +
    movies['cast'] + ' ' +
    movies['director']
)

movies[['title', 'text']].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['text'])

tfidf_matrix.shape
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
cosine_sim.shape
```

```python
# Title → index mapping (lowercase)
title_to_index = {title.lower(): idx for idx, title in enumerate(movies['title'])}

def recommend(movie_title, num_recommendations=5):
    key = movie_title.strip().lower()

    if key not in title_to_index:
        return f"❌ Movie not found: {movie_title}"

    idx = title_to_index[key]

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:num_recommendations+1]

    recommended_indices = [i for i, _ in sim_scores]
    return movies.iloc[recommended_indices]
```

```python
def show_recommendations(movie_title, num_recommendations=5):
    result = recommend(movie_title, num_recommendations)

    if isinstance(result, str):
        print(result)
        return

    print(f"\n🎬 Because you watched: {movie_title}\n")
    for i, row in result.iterrows():
        print(f"- {row['title']} ({row['release_year']})")
        print(f"  Rating: {row['rating']}")
        print(f"  Genres: {row['listed_in']}")
        print(f"  Description: {row['description'][:140]}...")
        print()
```

```python
show_recommendations("The Matrix", 5)
```
